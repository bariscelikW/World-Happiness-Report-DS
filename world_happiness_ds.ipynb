# Q1. Data Merging and Consistency Checking


1. Merge the datasets on PassengerId into a single DataFrame. How many rows are there in the merged dataset? **(3 Pts)**
2. Perform a consistency check for duplicated rows. After merging, check for any duplicated rows in the merged dataset and provide a summary of your findings. Provide the details (average, mean, etc.) of each feature in the merged DataFrame. **(3 Pts)**
3. Determine the survival rate for each passenger class and gender. Analyze how the survival rate varies among different classes (Pclass) and genders. **(3 Pts)**
4. Analyze the effect of age on survival probability. Determine whether age significantly affects the probability of survival. Use statistical techniques to explore this relationship.(use binning) **(4 Pts)**
5. Investigate the impact of family size on survival. Determine if family size (based on the number of siblings/spouses and parents/children aboard) has any correlation with survival. **(4 Pts)**
6. Determine the most common embarkation point among survivors. Find out which embarkation point (C, Q, or S) had the highest survival rate. **(4 Pts)**
7. Evaluate the correlation between Fare and Survival. Determine if there is a correlation between the fare paid and the survival status of the passengers. **(4 Pts)**

```python
#1
#merge datasets 
df1 = pd.read_csv('Data/passenger_demographics.csv')
df2 = pd.read_csv('Data/passenger_survival.csv')
df = pd.merge(df1, df2, on='PassengerId')

print(df.shape[0])
print(df)
```

```python
#2
#check for duplicate
duplicate_rows = df[df.duplicated()]
print(duplicate_rows)

# Summary statistics of the merged DataFrame
print(df.describe())

# Count of duplicate rows
print(duplicate_rows.shape[0])
```

```python
#3 Determine the survival rate for each passenger class and gender. Analyze how the survival rate varies 
# among different classes (Pclass) and genders.,

# Calculating survival rate for each class and gender
survival_rates = df.groupby(['Pclass', 'Sex'])['Survived'] 
print(survival_rates.describe())

# rate of alive decreasing towards from 1 to 3 and from man to woman
```

```python
#4 Analyze the effect of age on survival probability. Determine whether age significantly affects the probability of survival
#Use statistical techniques to explore this relationship.(use binning)

# Binning the age data into categories
bins = [0, 18, 35, 50, 80]
labels = ['0-18', '19-35', '36-50', '51-80']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)

# Calculating the survival rate for each age group
age_survival_rate = df.groupby('AgeGroup')['Survived'].mean() * 100

# survival rate is the highest in the early age 
print(age_survival_rate)
```

```python
#5 Investigate the impact of family size on survival. Determine if family size (based on the number of siblings/spouses
#and parents/children aboard) has any correlation with survival.

# Creating a new feature 'FamilySize' by adding 'SibSp' and 'Parch' to represent family size
df['FamilySize'] = df['SibSp'] + df['Parch']

# Calculating the survival rate for different family sizes
family_size_survival_rate = df.groupby('FamilySize')['Survived'].mean() * 100

# When I look at the infos, survival rate increasing until 3 after that decreasing
print(family_size_survival_rate)
```

```python
#6 Determine the most common embarkation point among survivors. Find out which embarkation point (C, Q, or S) 
#had the highest survival rate.

# Calculating the survival rate for each embarkation point
embarkation_survival_rate = df.groupby('Embarked')['Survived'].mean() * 100

# Finding the embarkation point with the highest survival rate
most_common_embarkation_survivor = embarkation_survival_rate.idxmax()

# Displaying the survival rates and the most common embarkation point among survivors
print(embarkation_survival_rate)
print(most_common_embarkation_survivor)
```

```python
#7 7. Evaluate the correlation between Fare and Survival. Determine if there is a correlation between the fare paid
#and the survival status of the passengers. **(4 Pts)**

# Calculating the correlation between 'Fare' and 'Survived'
fare_survival_correlation = df['Fare'].corr(df['Survived'])

# Displaying the correlation result
print(fare_survival_correlation)
# correlation result is positive, that means there is a correlation between fare and Survival
```

The Supermarket Sales Dataset is a comprehensive dataset designed to provide insights into the sales performance of a supermarket.

We want to store this data in a relational database, with the following table schemas:

# Tables

## 1. Branches Table

| Column   | Type  | Key      | Description                              |
|----------|-------|----------|------------------------------------------|
| BranchID | INTEGER  | PRIMARY  | Unique identifier for each branch        |
| Branch | TEXT  |            |    Name of the supermarket branch        |

## 2. Products Table

| Column      | Type  | Key      | Description                               |
|-------------|-------|----------|-------------------------------------------|
| ProductID | INTEGER  | PRIMARY  | Unique identifier for each product line   |
| ProductLine | TEXT  |          | Category of the product |



## 3. Orders Table

| Column        | Type  | Key      | Description                                    |
|---------------|-------|----------|------------------------------------------------|
| OrderID       | TEXT  | PRIMARY  | Unique identifier for each order               |
| BranchID      | INTEGER  |   FK       | Identifier for the branch of the order    |
| ProductID   | INTEGER  |      FK    | Identifier linking to Products table ordered                      |
| Quantity      | INTEGER |        | Number of units ordered                        |
| UnitPrice   | REAL  |          | Price per unit of the product             |Total         | REAL  |          | Total sales value for the order               |

| CustomerType  | TEXT  |          | Type of customer (e.g., Member, Normal)     |
| Gender        | TEXT  |          | Gender of the customer                        |



Summary of Table Structures

* Branches Table contains information about different supermarket branches.

* Products Table details the various products available for purchase.
* Orders Table records the transactions made, linking products and branches to specific purchases.


# **Questions** **(35 Pts)**

1. Read the csv file and create this DB **(10 Pts)**
2. Write a query to find the top 3 product lines with the highest total quantity sold, using a LIMIT statement. **(4 Pts)**
3. Find the total revenue generated for each branch by product line **(4 Pts)**
4. Find the total sales for each branch **(3 Pts)**
5. Determine the number of orders placed by each gender **(3 Pts)**
6. List the product lines sold in each branch and the number of orders **(4 Pts)**
7. Find the highest quantity sold for a single order and the corresponding product line **(4 Pts)**
8. Check for duplicated invoices **(3 Pts)**


```python
#1
# Load the CSV data into a pandas DataFrame
data = pd.read_csv("Data/supermarket_sales.csv")

# create and connect new db
conn = sqlite3.connect('supermarket_sales.db')
cursor = conn.cursor()


# Create the table structure based on the dataset
cursor.execute('''
CREATE TABLE IF NOT EXISTS sales (
    invoice_id TEXT,
    branch TEXT,
    city TEXT,
    customer_type TEXT,
    gender TEXT,
    product_line TEXT,
    unit_price REAL,
    quantity INTEGER,
    tax_5 REAL,
    total REAL,
    date TEXT,
    time TEXT,
    payment TEXT
)
''')

# Insert the CSV data into the SQLite table
for row in data.itertuples(index=False):
    cursor.execute('''
    INSERT INTO sales (invoice_id, branch, city, customer_type, gender, product_line, unit_price, quantity, tax_5, total, date, time, payment)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', row)
    
# Commit changes
conn.commit()
```

```python
#2
# Query to find the top 3 product lines with the highest total quantity sold
query_1 = '''
    SELECT product_line, SUM(quantity) AS total_quantity
    FROM sales
    GROUP BY product_line
    ORDER BY total_quantity DESC
    LIMIT 3;
'''
cursor.execute(query_1)
results_1 = cursor.fetchall()

for row in results_1:
    print(row)


```

```python
#3
# Find the total revenue generated for each branch by product line
query_2 = '''
    SELECT branch, product_line, SUM(total) AS total_revenue
    FROM sales
    GROUP BY branch, product_line
    ORDER BY branch, total_revenue DESC;
'''
cursor.execute(query_2)
results_2 = cursor.fetchall()

for row in results_2:
    print(row)
```

```python
#4
# Find the total sales for each branch
query_3 = '''
    SELECT branch, SUM(total) AS total_sales
    FROM sales
    GROUP BY branch
    ORDER BY total_sales DESC;
'''
cursor.execute(query_3)
results_3 = cursor.fetchall()

for row in results_3:
    print(row)
```

```python
#5
# Determine the number of orders placed by each gender
query_4 = '''
    SELECT gender, COUNT(*) AS num_orders
    FROM sales
    GROUP BY gender;
'''
cursor.execute(query_4)
results_4 = cursor.fetchall()

for row in results_4:
    print(row)
```

```python
#6
# List the product lines sold in each branch and the number of orders
query_5 = '''
    SELECT branch, product_line, COUNT(*) AS num_orders
    FROM sales
    GROUP BY branch, product_line
    ORDER BY branch, num_orders DESC;
'''
cursor.execute(query_5)
results_5 = cursor.fetchall()

for row in results_5:
    print(row)
```

```python
#7
# Find the highest quantity sold for a single order and the corresponding product line
query_6 = '''
    SELECT product_line, MAX(quantity) AS max_quantity
    FROM sales;
'''
cursor.execute(query_6)
results_6 = cursor.fetchall()

for row in results_6:
    print(row)
```

```python
#8

# Check duplicated invoices
query_7 = '''
    SELECT invoice_id, COUNT(*) AS count
    FROM sales
    GROUP BY invoice_id
    HAVING count > 1;
'''
cursor.execute(query_7)
results_7 = cursor.fetchall()

for row in results_7:
    print(row)

# Close the database connection
conn.close()
```
